```python
import pandas as pd
import numpy as np
```

```python
df1819Clean = pd.read_csv('Data20182019_No0_CleanFinal.csv')
len(df1819Clean)
```

```python
df1819Clean.head()
```

```python
df1819Clean.dtypes
```

```python
df_Code_Info = pd.read_excel('LookUp_Code_Info.xlsx')
df_Code_Info.head()
```

```python
df1819Clean = df1819Clean.merge(df_Code_Info,on="code")
df1819Clean.head()
```

```python
df1819Clean.dtypes
```

```python
df_Lat_Long = pd.read_excel('LookUp_Lat_Long.xlsx')
df_Lat_Long.head()
```

```python
df1819Clean = df1819Clean.merge(df_Lat_Long,on="station")
df1819Clean.head()
```

```python
df1819Clean.dtypes
```

```python
df_Line_Name = pd.read_excel('LookUp_Line.xlsx')
df_Line_Name.head()
```

```python
df1819Clean = df1819Clean.merge(df_Line_Name,on="line")
df1819Clean.head()
```

```python
import datetime
tempdate = pd.to_datetime(df1819Clean['date'])
df1819Clean['month'] = tempdate.dt.month_name(locale = 'English')
df1819Clean.head()
```

```python
#df1819Clean['month_number'] = pd.DatetimeIndex(df1819Clean['date']).month
df1819Clean['month_number'] = tempdate.dt.month
df1819Clean['year'] = tempdate.dt.year
df1819Clean.head()
```

```python
#other method
#df1819Clean['year'] = pd.DatetimeIndex(df1819Clean['date']).year
```

```python
temphour = pd.to_datetime(df1819Clean['time'])
df1819Clean['hour'] = temphour.dt.hour
df1819Clean.head()
```

```python
# Create the bins in which Data will be held 
bins = [0, 5, 9, 12, 15, 18, 23]

# Create the names for the four bins
group_names = ["9PM-1:30AM", "5-9AM", "9AM-12PM", "12-3PM", "3-6PM","6-9PM"]
```

```python
df1819Clean["time_range"] = pd.cut(df1819Clean["hour"], bins, labels=group_names)
df1819Clean.head()
```

```python
df1819Clean.shape
```

```python
df1819Clean['id'] = df1819Clean.index +1
```

```python
df1819Final = df1819Clean[['id', 'date', 'time', 'day', 'station', 'code', 'min_delay', 'min_gap',
       'bound', 'line', 'vehicle', 'code_info', 'latitude', 'longitude',
       'line_name', 'month', 'time_range', 'month_number', 'hour', 'year']]
df1819Final.head()
```

```python
df1819Final.to_csv('ttc_subway_delay_2018_2019_from_pandas.csv', index=False)
```
